```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import sklearn.svm
%matplotlib inline

# Utility functions
def plotDataset(X, y):
    y = y.reshape(-1)
    plt.plot(X[y!=1,0],X[y!=1,1],'bo',X[y==1,0],X[y==1,1],'rx')

def plotBoundary(decision_function, X, margin=False, N=100):
    x1 = np.linspace(X[:,0].min(), X[:,0].max(), N)
    x2 = np.linspace(X[:,1].min(), X[:,1].max(), N)
    xx1, xx2 = np.meshgrid(x1, x2)
    X = np.column_stack((xx1.reshape(-1),xx2.reshape(-1)))
    z = decision_function(X).reshape(xx1.shape)
    plt.contourf(xx1,xx2,np.sign(z),cmap=plt.cm.bwr,vmin=-3,vmax=3)
    plt.contour(xx1,xx2,z,cmap=plt.cm.gray,vmin=np.inf,vmax=np.inf,levels=[0])
    if margin:
        plt.contourf(xx1,xx2,z,cmap=plt.cm.bwr,vmin=-3,vmax=3,levels=[-1,0,1])

def plotCircles(X):
    if X.shape[1]>0:
        plt.plot(X[:,0],X[:,1],'go',mfc='none',ms=12)
```

## Kernel Gaussiano:

```python
#def gaussianKernel(X1,X2, σ=2):
def gaussianKernel(X1,X2):
    global σ
    matrix = np.zeros((X1.shape[0],X2.shape[0]))
    
    for i, x1 in enumerate(X1):
        for j, x2 in enumerate(X2):
            x1 = x1.flatten()
            x2 = x2.flatten()
            matrix[i, j] = np.exp(-np.sum(np.power((x1-x2),2))/(2*(σ**2)))

           #matrix[i,j] = np.exp(- (   /(2*σ**2)))
    return matrix
```

```python
def drange(start, stop, step):
    r = start
    while r < stop:
        yield r
        r += step
```

```python
σ = 1
for σ in drange(0.1, 2, 0.1):
    model = sklearn.svm.SVC(C=1, kernel=gaussianKernel)
    model.fit(X,y)
    print('Accuracy:',model.score(X,y), "Para um σ = ", σ)
    plotBoundary(model.decision_function,X,margin=True,N=30)
    plotDataset(X,y)
    plotCircles(model.support_vectors_)
    plt.show()
```

# Terceiro Dataset
Repetindo os mesmos procedimentos

```python
data3 = loadmat('data/ex6data3.mat')

y3 = data3['y']
X3 = data3['X']

Xval, yval = data3['Xval'], data3['yval']

print('X3:', X3.shape)
print('y3:', y3.shape)
```

```python
plotDataset(X3,y3)
```

#### No exemplo do github pede-se uma rotina para avaliar o melhor C, juntamente ao $\sigma$, correspondente ao melhor par. Sendo assim obtendo-se a melhor acurácia

```python
C_vec = (0.01, 0.03, 0.1, 0.3, 1., 3., 10., 30.)
σ_vec = C_vec

best_C = 0
best_σ = 0
best_accuracy = 0
best_gamma = 0

for C_val in C_vec:
    for σ_val in σ_vec:
        gamma = 1/(σ_val**2)
        model3 = sklearn.svm.SVC(C=C_val, kernel='rbf',gamma = gamma)
        model3.fit(X3,y3)
        accuracy = model3.score(Xval,yval)
        
        plotBoundary(model3.decision_function,X3,margin=True)
        plotDataset(X3,y3)
        plotCircles(model3.support_vectors_)
        plt.show()
        
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_C = C_val
            best_σ = σ_val
            best_gamma = 1/(best_σ**2)

print("Melhor C = ", best_C)
print("Melhor σ = ", best_σ)
print("Melhor γ = ", best_gamma)
print("Melhor Acurácia = ", best_accuracy)

```

```python
gamma = best_gamma
model3 = sklearn.svm.SVC(C=best_C, kernel=gaussianKernel,gamma = gamma)
model3.fit(X3,y3)
        
```

```python
plotBoundary(model3.decision_function,X3,margin=True)
plotDataset(X3,y3)
#plotDataset(Xval,yval)
plotCircles(model3.support_vectors_)
```
